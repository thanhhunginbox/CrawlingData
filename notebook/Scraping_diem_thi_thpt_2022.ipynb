```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
list_data = []
list_data_failed = []
```

```python
n = 50014843
max = 50015099
while n <= max:
    url = 'https://vietnamnet.vn/giao-duc/diem-thi/tra-cuu-diem-thi-tot-nghiep-thpt/2022/' + str(n) + '.html'
    reponse = requests.get(url=url, verify=False)

    if reponse.status_code != 200:
        list_data_failed.append(n)
        n = n+1
        continue

    soup = BeautifulSoup(reponse.text, 'html.parser')
    list_td = soup.tbody.find_all('td')

    dict = {}
    dict['SBD'] = str(n)
    for i, td_tag in enumerate(list_td):    
        if i % 2 != 0:
            continue
        dict[td_tag.string] = float(list_td[i+1].string)

    list_data.append(dict)
    n = n+1
```

```python
print(len(list_data))
print(len(list_data_failed))
```

```python
print(list_data[len(list_data)-1])
# print(list_data)
```

```python
df = pd.DataFrame(list_data)
df_failed = pd.DataFrame(list_data_failed)
```

```python
df.to_csv(f'data_{n}_{max}.csv')
df_failed.to_csv('failed4')
```
